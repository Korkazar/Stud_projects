```python
#зададим функцию графического анализа
def graph_builder(data, to_plot, to_hue, graph_row, graph_column, width, height):
    fig, axes = plt.subplots(graph_row, graph_column, figsize=(width, height)) #количество графиков задаётся переменными
    #необходимые счётчики для координат графиков в фигуре
    c=0
    t=0
    
    for plot in to_plot: #Перебор столбцов из полученного списка
        if c>=graph_column: #Проверка, следует ли менять ряд
            t+=1
            c=0
        #Построение графика - присваивание положения, передача данных и т.д. Для категориальных данных строится столбчатый график
        if data.loc[:,plot].dtype == 'object':
            sns.barplot(
                y = data.groupby(plot)['id'].count().reset_index()[plot],
                x = data.groupby(plot)['id'].count().reset_index()['id'],
                ax = axes[t,c],
                hue = data.groupby(plot)['id'].count().reset_index()[plot]
                )
       
            axes[t,c].set_xlabel(f'Распределение в  {plot}')
            axes[t,c].set_ylabel('Сотрудников')
        else:
            if len(data.loc[:,plot].unique())>20:
                sns.histplot(data, x=plot, bins = 15, ax = axes[t,c], hue = to_hue) 
        
                axes[t,c].set_xlabel(f'Распределение в  {plot}')
                axes[t,c].set_ylabel('Сотрудников')
            else:
                sns.countplot(data, x=plot, ax = axes[t,c], hue = to_hue) 
        
                axes[t,c].set_xlabel(f'Распределение в  {plot}')
                axes[t,c].set_ylabel('Сотрудников')
            c+=1
            if c>=graph_column: 
                t+=1
                c=0
            sns.boxplot(data, x=plot, ax = axes[t,c], hue = to_hue)  
            axes[t,c].set_xlabel(f'Квартили и выбросы в  {plot}')
            axes[t,c].set_ylabel('')
    
        c+=1
    
    fig.tight_layout()      
    plt.show()
```

### Выбор модели машинного обучения

Первоначально - создадим пайплайн предобработки данных, обрабатывающий пропуски через SimpleImputer, кодирующий и масштабирующий признаки.  
Категориальные признаки без рангов обработаем черех OneHotEncoder, ранговые - OrdinalEncoder.  
Масштабирование данных произведём при помощи MinMaxScaler.  

#### Пайплайн предобработки данных

Составим списки столбцов в данных, передаваемые пайплайну.

```python
ohe_columns=['dept', 'last_year_promo', 'last_year_violations']
ord_columns=['level', 'workload']

num_columns=['employment_years','supervisor_evaluation', 'salary']
```

```python

#пайплан кодирования неранговых категорий
ohe_pipe = Pipeline(
    [
        ('simp_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore',  sparse_output=False))
    ])

#пайплан кодирования ранговых категорий
ord_pipe = Pipeline(
    [
        (
            'simp_first', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'sinior'],
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
        (
            'simp_last', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)


# общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ]
)
```

Проверим работу пайплайна.  
Для этого - разметим данные для удобства дальнейшей работы.

#### Метрика SMAPE

Согласно проектному заданию, подготовим критерий выбора лучшей модели - метрику SMAPE.

```python
#Спасибо преподавателю за более изящное решение
def smape(y_true, y_pred):
    error = np.abs(y_true - y_pred)
    scale = (np.abs(y_true) + np.abs(y_pred)) / 2
    return np.mean(error / scale) * 100

scorer = make_scorer(smape, greater_is_better = False)
```

Успешным значением считается метрика SMAPE < 15.  
  
Перейдём к выбору модели.

Для модели дерева решений будет осуществляться подбор гиперпараметров.  
Подбор гиперпарметров осуществляется методом байесовской оптимизации, реализованный в библиотеке Optuna.  
Подбор и обучения моделей будет осуществляться по отдельности, для улучшения совместимости с библиотекой Optuna.

#### Портрет уволившегося сотрудника

Рассмотрим имеющиеся данные, разделив их по факту увольнения.

```python
def graph_builder_normal(data, to_plot, to_hue, graph_row, graph_column, width, height):
    fig, axes = plt.subplots(graph_row, graph_column, figsize=(width, height)) #количество графиков задаётся переменными
    #необходимые счётчики для координат графиков в фигуре
    c=0
    t=0
    
    for plot in to_plot: #Перебор столбцов из полученного списка
        if c>=graph_column: #Проверка, следует ли менять ряд
            t+=1
            c=0
        #Построение графика - присваивание положения, передача данных и т.д. Для категориальных данных строится столбчатый график
        if data.loc[:,plot].dtype == 'object':
            sns.barplot(
                y = data[plot].value_counts(normalize=True).reset_index()[plot],
                x = data[plot].value_counts(normalize=True).reset_index()['index'],
                ax = axes[t,c],
                hue = data[plot].value_counts(normalize=True).reset_index()['index']
                )
       
            axes[t,c].set_xlabel(f'Распределение в  {plot}')
            axes[t,c].set_ylabel('Доля сотрудников')
        else:
            if len(data.loc[:,plot].unique())>20:
                sns.histplot(data, x=plot, bins = 15, ax = axes[t,c], hue = to_hue, stat='density', common_norm=False) 
        
                axes[t,c].set_xlabel(f'Распределение в  {plot}')
                axes[t,c].set_ylabel('Плотность распределения')
            else:
                sns.countplot(data, x=plot, ax = axes[t,c], hue = to_hue) 
        
                axes[t,c].set_xlabel(f'Распределение в  {plot}')
                axes[t,c].set_ylabel('Сотрудников')
            c+=1
            if c>=graph_column: 
                t+=1
                c=0
            sns.boxplot(data, x=plot, ax = axes[t,c], hue = to_hue)  
            axes[t,c].set_xlabel(f'Квартили и выбросы в  {plot}')
            axes[t,c].set_ylabel('')
    
        c+=1
    
    fig.tight_layout()      
    plt.show()
```

```python
to_plot = data_train_quit.loc[:, ((data_train_quit.columns != 'id')&(data_train_quit.columns != 'quit'))].select_dtypes(include='number').columns

graph_builder_normal(data_train_quit, to_plot, data_train_quit['quit'], 3, 2, 10, 10)
```

### Выбор модели 

#### Разметка данных

Сформруем тренировочную выборку для дальнейшего моделирования.  
Для этого, заново зададим все списки столбцов и переопределим пайплайн.  
Столбец level исключим как возможный источник переобучения.

```python
ohe_columns=['dept', 'last_year_promo', 'last_year_violations']
ord_columns=['workload']
num_columns=['employment_years','supervisor_evaluation', 'salary', 'job_satisfaction_rate']
```

```python
ord_pipe2 = Pipeline(
    [
        (
            'simp_first', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['low', 'medium', 'high'],
                ], 
                handle_unknown='use_encoded_value',
                unknown_value=np.nan
            )
        ),
        (
            'simp_last', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)


data_preprocessor2 = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe2, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ]
)
```

```python
X_train_new = data_train_quit.drop(['id','level'], axis=1)
print('Вновь выявленных дубликатов в данных:',X_train_new.duplicated().sum())
print('Значений в датасете до удаления дубликатов:',len(X_train_new))
X_train_new = X_train_new.drop_duplicates()
print('Значений в датасете после удаления дубликатов:',len(X_train_new))
y_train_new = X_train_new['quit']
X_train_new = X_train_new.drop(['quit'], axis=1)


```

```python
label_encoder = LabelEncoder()


y_train_new_processed = label_encoder.fit_transform(y_train_new)
len(y_train_new_processed)
```

```python
X_train_processed_new= data_preprocessor2.fit_transform(X_train_new)

X_train_processed_new= pd.DataFrame(X_train_processed_new)
X_train_processed_new.columns = data_preprocessor2.get_feature_names_out()
display(X_train_processed_new)
X_train_processed_new.describe()
```

Для моделирования будут использоваться модели:
- DecisionTreeClassifier;
- KNeighborsClassifier;
- LogisticRegression.
  
Подбор гиперпараметров будет осуществляться при помощи байесовской оптимизации.  
Зададим распределения гиперпараметров для каждой модели.

```python
param_tree = {
    #'max_depth': distributions.IntDistribution(2, 6),
    'min_samples_leaf': distributions.IntDistribution(1, 10),
    'max_features': distributions.IntDistribution(2, 15)
}

param_knn = {
    'n_neighbors': distributions.IntDistribution(2, 15),
    'metric':distributions.CategoricalDistribution(choices=['minkowski', 'cityblock','cosine'])
}

    
param_lr = {
    'C': distributions.IntDistribution(1, 20),
    'penalty':distributions.CategoricalDistribution(choices=['l1','l2'])
}

```

#### Подбор гиперпараметров модели дерева решений

Осуществим подбор и обучение модели.

```python
oscv2 = OptunaSearchCV(
    DecisionTreeClassifier(max_depth = None, random_state=RANDOM_STATE),
    param_tree,
    scoring='roc_auc',
    n_jobs=None,
    cv=5,
    n_trials=100,
    random_state=RANDOM_STATE
) 
```

```python
oscv2.fit(X_train_processed_new, y_train_new_processed)
```

```python
print('Лучшие параметры модели: ', oscv2.best_estimator_)
print('Лучшая метрика: ',round(oscv2.best_score_,2))
```

Метрика близка к целевым значениям, обучение прошло успешно.

#### Подбор гиперпараметров модели ближайших соседей

```python
oscv3 = OptunaSearchCV(
    KNeighborsClassifier(),
    param_knn,
    scoring='roc_auc',
    n_jobs=None,
    cv=5,
    n_trials=100,
    random_state=RANDOM_STATE
) 
```

```python
oscv3.fit(X_train_processed_new, y_train_new_processed)
```

```python
print('Лучшие параметры модели: ', oscv3.best_estimator_)
print('Лучшая метрика: ',round(oscv3.best_score_,2))
```

Метрика близка к целевым значениям, обучение прошло успешно.

#### Подбор гиперпараметров модели логистической регресии

```python
oscv4 = OptunaSearchCV(
    LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            ),
    param_lr,
    scoring='roc_auc',
    n_jobs=None,
    cv=5,
    n_trials=200,
    random_state=RANDOM_STATE
) 
```

```python
oscv4.fit(X_train_processed_new, y_train_new_processed)
```

Метрика близка к целевым значениям, обучение завершилось успешно.
